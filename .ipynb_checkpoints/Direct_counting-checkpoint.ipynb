```python
import time
import numpy as np
import matplotlib.pyplot as plt
from multiprocessing import Pool
import itertools
```

```python
# The trajectory data contains two columns corresponding to x and y from the MD simulation 
# of a 3-state potential at beta=1/kT=10.
# Read the data file and visualize along x-direction:
beta=10
infile = 'xvyw1beta10.0gammax1.0gammay1.0epsln1.0sgma1.0A1.0x01.122w0.8B0.15a1.0_h0.01_mix1.txt'
data_x, data_y = np.loadtxt(infile, unpack=True, usecols=(0,1))

# Make a contour plot: Show the 3-state structure
hist=np.histogram2d(data_x, data_y, bins=25)
prob=hist[0].T/np.sum(hist[0])
x_axis=0.5*(hist[1][1:]+hist[1][:-1])
y_axis=0.5*(hist[2][1:]+hist[2][:-1])
freeE=(-1/beta)*np.log(prob+1e-7)

plt.contourf(x_axis, y_axis, freeE)
plt.colorbar()
plt.tick_params(axis='both', which='major', labelsize=12)
plt.xlabel('x', size=12)
plt.ylabel('y', size=12).set_rotation(0)
plt.show()

# Trajectory along x-direction:
fig, ax = plt.subplots(figsize=(12,5), nrows=1, ncols=1)
ax.plot(data_x)
ax.set_title('actual')
ax.tick_params(axis='both', which='major', labelsize=12)
ax.set_xlabel('step', size=12)
ax.set_ylabel('x', size=12).set_rotation(0)
plt.show()
```

```python
# To easily distinguish the states, we can do running average through convolution or digitization.
def running_mean(x, N):
    """Use convolution to do running average."""
    return np.convolve(x, np.ones((N,))/N, mode='valid')

# Running average:
data_x_hat = running_mean(data_x, 100)
# Digitization:
bins=np.linspace(-5, 5, 25)
idx_x=np.digitize(data_x, bins)

# Smoothen trajectory along x-direction:
fig, ax = plt.subplots(figsize=(12,10), nrows=2, ncols=1)
ax[0].plot(data_x_hat)
ax[0].tick_params(axis='both', which='major', labelsize=12)
ax[0].set_xlabel('step', size=12)
ax[0].set_ylabel('x', size=12).set_rotation(0)

ax[1].plot(idx_x)
ax[1].tick_params(axis='both', which='major', labelsize=12)
ax[1].set_xlabel('step', size=12)
ax[1].set_ylabel('idx', size=12).set_rotation(0)

plt.show()
```

```python
# The 3 states are located at x=(1.5, 0, -1.5) or correspondingly at idx~=(16, 12.5, 9).
# To find the number of transitions between different states using only trajectory 
# along x, we need to define a commit step/commitment to slide along trajectory. It is 
# then defined that the transition will be count only when x/idx stay in a state for more 
# than the commit step. A direct version of doing this is as follow:

class Transit:

    def __init__(self, traj, X2label, min_jump):
        self.traj=traj
        self.X2label=X2label # {state1:'A', state2:'B', state2:'C'}
        self.X=list(self.X2label)

        self.min_jump=min_jump

    def find_nearest(self, key_arr, target):
        """key_arr: array-like, storing keys.
           target: the value which we want to be closest to."""
        idx=np.abs(key_arr-target).argmin()
        return key_arr[idx]

    def countTransit(self, window):
        """Count the number of transitions."""
        start=self.find_nearest(self.X, self.traj[0])
        count={}
    
        for i in range(0, len(self.traj)-window):
            subarr=self.traj[i:i+window] # window = size of subarray = commit time
            nearest=self.find_nearest(subarr, start)
            if np.abs(nearest-start)>self.min_jump:
                prev_start=start
                start=self.find_nearest(self.X, np.mean(subarr))
                if prev_start != start:
                    key='in {} to {}'.format(self.X2label[prev_start], self.X2label[start])
                    # update count
                    if count.get(key) is None: count[key]=1
                    else: count[key]+=1
        return count
  
startT=time.time()

transit1=Transit(data_x, {1.5:'A', 0.0:'B', -1.5:'C'}, 0.75)
transit2=Transit(data_x_hat, {1.5:'A', 0.0:'B', -1.5:'C'}, 0.75)
transit3=Transit(idx_x, {16:'A', 12.5:'B', 9:'C'}, 1.75)

window_list = np.arange(5,250, 5)
with Pool(4) as p:
    x1 = p.map(transit1.countTransit, window_list)
    x2 = p.map(transit2.countTransit, window_list)
    x3 = p.map(transit3.countTransit, window_list)
    
print(time.time()-startT, 's')
```

```python
# Here we extract only the number of counts from A to B.
def extract_count(x, key):
    x1=[]
    for xi in x:
        if xi.get(key)==None: x1.append(0)
        else: x1.append(xi[key])
    return x1

count1 = extract_count(x1, 'in A to B')
count2 = extract_count(x2, 'in A to B')
count3 = extract_count(x3, 'in A to B')

fig, ax = plt.subplots(figsize=(8,6), nrows=1, ncols=1)
ax.scatter(window_list, count1, label='actual')
ax.scatter(window_list, count2, label='smoothen')
ax.scatter(window_list, count3, marker='^', label='digitized')

ax.tick_params(axis='both', which='major', labelsize=12)
ax.set_xlabel('commit time', size=16)
ax.set_ylabel('count', size=16)
ax.grid(which='both')
ax.legend()
plt.show()
```

The first platau the profile reach from the left correpsonds to the expected number of transitions between basins/states A and B. Note the actual one means raw trajectory.

```python
# A faster version:
class NewTransit:
    
    def __init__(self, traj, X2label, radius):
        self.traj=traj
        self.X2label=X2label
        self.radius=radius
         
    def classify(self, traj0):
        """Convert every element in traj0 to centers."""
        traj=np.copy(traj0).astype(np.float64)
        radius=self.radius
        center=list(self.X2label)
        for ci in center:
            traj[np.where((traj - ci)**2 < radius**2)]=ci
        idx_not_center=np.where(np.in1d(traj, center, invert=True))
        traj = np.delete(traj, idx_not_center)
        return traj

    def countTransit(self, window):
        """Count the number of transitions."""
        input_x1=self.classify(self.traj)

        start=input_x1[0]
        sum_rest=0
        rest_step={}
        count={}
        for li in itertools.permutations(self.X2label.values(),2):
            key='in {} to {}'.format(li[0], li[1])
            rest_step[key]=[]
            count[key]=0

        i=0
        mix=False
        while i < len(input_x1)-window:
    
            subarr=input_x1[i:i+window] # window = size of subarray = commit time
    
            if len(set(subarr))==1 and mix==False: # all elements in the subarr are the same.
                sum_rest+=window
                i+=window
    
            elif len(set(subarr)) > 1:
                if start in subarr: last_elem=start
                else: last_elem=subarr[-1]
                j=0
                while subarr[-j-1] != last_elem: j+=1
                temp_transit=window-j # temporary/fake transitions during this interval.
        
                sum_rest+=temp_transit
                i+=temp_transit
                mix=True
        
            elif len(set(subarr)) == 1 and mix == True:
                prev_start=start
                start=subarr[0]
                if prev_start != start:
                    if start == last_elem: sum_rest+=j+1
                    key='in {} to {}'.format(self.X2label[prev_start], self.X2label[start])
                    rest_step[key].append(sum_rest) # update rest step
                    sum_rest=0 # reset sum_rest
                    count[key]+=1 # update count
        
                sum_rest+=window
                i+=window
                mix=False

        return count
        
startT=time.time()

transit1=NewTransit(data_x, {1.5:'A', 0.0:'B', -1.5:'C'}, 0.75)
transit2=NewTransit(data_x_hat, {1.5:'A', 0.0:'B', -1.5:'C'}, 0.75)
transit3=NewTransit(idx_x, {16:'A', 12.5:'B', 9:'C'}, 1.75)

window_list = np.arange(5,250, 5)
with Pool(4) as p:
    x1 = p.map(transit1.countTransit, window_list)
    x2 = p.map(transit2.countTransit, window_list)
    x3 = p.map(transit3.countTransit, window_list)
    
print(time.time()-startT, 's')
```

```python
# Here we extract only the number of counts from A to B.
def extract_count(x, key):
    x1=[]
    for xi in x:
        if xi.get(key)==None: x1.append(0)
        else: x1.append(xi[key])
    return x1

count1 = extract_count(x1, 'in A to B')
count2 = extract_count(x2, 'in A to B')
count3 = extract_count(x3, 'in A to B')

fig, ax = plt.subplots(figsize=(8,6), nrows=1, ncols=1)
ax.scatter(window_list, count1, label='actual')
ax.scatter(window_list, count2, label='smoothen')
ax.scatter(window_list, count3, marker='^', label='digitized')

ax.tick_params(axis='both', which='major', labelsize=12)
ax.set_xlabel('commit time', size=16)
ax.set_ylabel('count', size=16)
ax.grid(which='both')
ax.legend()
plt.show()
```

The platau is at around 10. Therefore, we expect there should be 10 transitions.
